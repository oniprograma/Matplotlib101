```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y = [10,3,5,8,1,4]
plt.scatter(x,y,label="puntos",color="r",s=25,marker="o")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.title("Mi primer grafica de dispersión \nDatos crudos")
```

```python

# height and weight data
height = [167, 175, 170, 186, 190, 188, 158, 169, 183, 180]
weight = [65, 70, 72, 80, 86, 94, 50, 58, 78, 85]

# plot a scatter plot
plt.scatter(weight, height)
# set axis lables
plt.xlabel("Weight (Kg)")
plt.ylabel("Height (cm)")
# set chart title
plt.title("Height v/s Weight")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y = [10,3,5,8,1,4]

nuevo_x = [i+1 for i in x] #agregando 1 a la lista "x"
nuevo_y = [i+1 for i in y] #agregando 1 a la lista "y"

plt.scatter(x,y,label="estrellas",color="b",s=200,marker="*") 
plt.scatter(nuevo_x,nuevo_y,label="triangulos",color="r",s=200,marker="v")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.title("Gráfica de dispersión")
```

# Iris

```python
import pandas as pd
import matplotlib.pyplot as plt

iris_data = pd.read_csv("/content/Iris.csv")

versicolor = iris_data[iris_data.Species == "Iris-versicolor"]
setosa = iris_data[iris_data.Species == "Iris-setosa"]
virginica = iris_data[iris_data.Species == "Iris-virginica"]

fig, ax = plt.subplots()
fig.set_size_inches(8, 8)


ax.scatter(versicolor["PetalLengthCm"], versicolor["PetalWidthCm"], label="versicolor")
ax.scatter(setosa["PetalLengthCm"], setosa["PetalWidthCm"], label="setosa")
ax.scatter(virginica["PetalLengthCm"], virginica["PetalWidthCm"], label="virginica")


ax.set_xlabel("Longitud del petalo")
ax.set_ylabel("Ancho del petalo")
ax.set_title("Tamaño de los las flores")
ax.legend()
```
