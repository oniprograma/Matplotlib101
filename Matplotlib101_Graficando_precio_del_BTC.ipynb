Seleccionamos una ventana de tiempo y descargamos los datos

```python
import matplotlib.pyplot as plt

import pandas as pd

datos = pd.read_csv('/content/BTC-USD.csv')
datos.head()
```

```python
fechas = datos['Date'].to_list()
apertura = datos['Open'].to_list()
```

```python
plt.plot(fechas,apertura,'-', label='Precio')
 
plt.xlabel('tiempo')
plt.ylabel('Precio')
plt.title('Cambio del Bitcoin')
plt.legend()
plt.show()
```

```python
import matplotlib.dates as mdates
import numpy as np

fig = plt.figure()
ax1 = plt.subplot2grid((1,1), (0,0))

ax1.plot(fechas,apertura, '-', label='Precio')

ax1.set_xticks(np.arange(0, len(fechas), 90))
plt.xlabel('Minuto')
plt.ylabel('Precio')
plt.title('Cambio del BTC')
plt.legend()

plt.show()
```
