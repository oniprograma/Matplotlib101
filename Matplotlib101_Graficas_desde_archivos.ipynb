```python
import matplotlib.pyplot as plt
import numpy as np

x, y = np.loadtxt('/content/ejemplo.txt', delimiter=',', unpack=True)

plt.plot(x,y, label='Gráfica')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfica generada \n desde el archivo')
plt.legend()
plt.show()
```

```python
print(x,y)
```

```python
import csv


x = []
y = []

with open('/content/ETH-USD.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    for row in csv_reader:
      try:
        x.append(row[0])
        y.append(row[1])
      except:
        pass

plt.plot(x,y, label='Gráfica')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfica generada \n desde el archivo')

plt.legend()
```
