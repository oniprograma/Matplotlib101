```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#importamos los datos
datos = pd.read_csv('/content/BTC-USD.csv')
```

```python
#todas las fechas
prueba = datos
fechas = datos['Date'].to_list()
#primer trimestre
# prueba = datos.iloc[:90]
# fechas = datos['Date'].to_list()[:90] #fechas de los primeros tres meses
```

```python
fig = plt.figure()
ax1 = plt.subplot2grid((1,1), (0,0))

ancho = 0.3 #ancho rectangulo
ancho2 = 0.1 #ancho mecha

#tipo de movimiento
alcista = prueba[prueba.Close>=prueba.Open]
bajista = prueba[prueba.Close<prueba.Open]

#colores
col1 = 'green'
col2 = 'red'
#Imprimir precios alcistas
ax1.bar(alcista.index,alcista.Close-alcista.Open,ancho,bottom=alcista.Open,
        color=col1) #rectangulo
ax1.bar(alcista.index,alcista.High-alcista.Close,ancho2,bottom=alcista.Close,color=col1) #sombra superior
ax1.bar(alcista.index,alcista.Low-alcista.Open,ancho2,bottom=alcista.Open,color=col1) #sombra inferior

#Imprimir precios bajistas
ax1.bar(bajista.index,bajista.Close-bajista.Open,ancho,bottom=bajista.Open,color=col2) #rectangulo
ax1.bar(bajista.index,bajista.High-bajista.Open,ancho2,bottom=bajista.Open,color=col2) #sombra superior
ax1.bar(bajista.index,bajista.Low-bajista.Close,ancho2,bottom=bajista.Close,color=col2) #sombra inferior

#todas las fechas
ax1.set_xticks(np.arange(0, len(fechas), 90))
ax1.set_xticklabels(fechas[::30])

#Solo el primer trimestre
# ax1.set_xticks(np.arange(0, len(fechas), 30))
# ax1.set_xticklabels(fechas[::30])



#Modificando el ángulo de las etiquetas en X
for etiqueta in ax1.xaxis.get_ticklabels():
    etiqueta.set_rotation(45)

fig.set_size_inches(18.5, 10.5) #tamaño de la imágen

plt.xlabel('Fechas')
plt.ylabel('Precio')
plt.title('Grafica de velas del primer trimestre\nPrecio BTC')
plt.legend()
```
